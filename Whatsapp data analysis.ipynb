```python
import numpy as np
import pandas as pd 
import re
import dateutil
import matplotlib.pyplot as plt
import seaborn as sb
from collections import Counter

with open('WhatsApp Chat with YABATECH 201819 ASPIRANT.txt','r',encoding='utf8') as file:
    data_file=file.read()
    data_file
    print(data_file)
```

```python
#Cleaning data with regular expression
#data_found = re.findall('(\d+/\d+/\d+,\s+\d+\:\d+)\s+\-\s(\\u202a\+\d+\s\d+\s\d+\s\d+\u202c|\w+\s\w+\s\w+|\w+\s\w+|\w+)\:(.*)', data_file)
data_found = re.findall('(\d+/\d+/\d+),\s+(\d+\:\d+\s+\w+)\s+\-\s(\w+\s\w+\s\w+|\w+\s\w+|\w+)\:(.*)', data_file)
data_found
```

```python
#Converting data_found to pandas dataframe
df=pd.DataFrame(data_found,columns=['Date','Time','Sender','Message'])
df.head()
```
